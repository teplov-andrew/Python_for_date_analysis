1. Как зовут владельца, у недвижимости которого за отчетный период была наибольшая разница между максимальной и минимальной ценой?

```python
# очищаем столбец с ценой и превращаем его в float
data_listings["price_clear"] = data_listings['price'].str.replace('$', '')
data_listings["price_clear"] = data_listings['price_clear'].str.replace(',', '').astype(float)
data_listings["price_clear"]
```

```python
# некоторые владельцы сдают комнату несколько раз
data_listings.shape[0] - len(set(data_listings["host_name"]))
```

```python
# вычисляем разницу между максимальной и минимальной ценой
data_listings["price_max_min_razn"] = data_listings.groupby("host_name")["price_clear"].transform(lambda x: x.max() - x.min())
data_listings["price_max_min_razn"]
```

```python
# вычисляем имя владельца у которого за отчетный период была наибольшая разница между максимальной и минимальной ценой
set(data_listings[data_listings["price_max_min_razn"] == data_listings["price_max_min_razn"].max()]["host_name"])
```

4. Давайте рассмотрим кейс, в котором имена хозяина недвижимости и гостя совпадают. Как зовут хозяйку, у которой было в качестве гостей больше всего тёзок? Что все её гости отмечали в качестве плюсов сдаваемой недвижимости?

```python
data_merged = pd.merge(data_listings, data_reviews, left_on='host_name', right_on='reviewer_name')
data_merged.head(2)
```

```python
data_merged[data_merged["reviewer_name"] == data_merged["host_name"]]["host_name"].value_counts()
```

```python
# поиск хозяйку с наибольшим количеством отзывов от тесок
best_host = data_merged[data_merged["reviewer_name"] == data_merged["host_name"]]["host_name"].value_counts().index[3]
best_host
```

```python
data_merged[data_merged["reviewer_name"] == data_merged["host_name"]]["host_name"]
```

```python
# положительный отзывы
data_merged[data_merged[data_merged["reviewer_name"] == data_merged["host_name"]]["host_name"] == best_host]["comments"].to_list()[:4]
```
