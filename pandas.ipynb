```python
data_films = data_netflix[data_netflix["type"]=="Movie"]
```

```python
data_top_10_contry_movie = data_films.groupby(["country", "type"]).size().unstack().sort_values("Movie", ascending=False).head(10)
data_top_10_contry_movie
```

```python
data_mivie_years = {country: data_films[data_films['country'] == country]['release_year'] for country in data_top_10_contry_movie.index}
```

```python
fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(18, 12), sharex=True, sharey=True)
fig.suptitle('Распределение фильмов по годам выпуска в разных странах')

for ax, country in zip(axes.flatten(), data_top_10_contry_movie.index):
    sns.kdeplot(data_mivie_years[country], fill=True, color="g", alpha=.7,ax=ax)
    ax.set_title(country)
    ax.set_xlabel('Год выпуска')
    ax.set_ylabel('Количество фильмов')
    ax.grid(True)

plt.show()
```

```python
data_fifa = pd.read_csv("/content/drive/MyDrive/hw3-2/FIFA/CompleteDataset.csv")
data_fifa.head()
```

```python
data_fifa["wage_clean"] = data_fifa["Wage"].str.replace("€","")
data_fifa["wage_clean"] = data_fifa["wage_clean"].str.replace("K","").astype(int)*1000
data_fifa.head()
```

```python
data_fifa_top_1000_wage = data_fifa.sort_values("wage_clean", ascending=False)
data_fifa_top_1000_wage.head()
```

```python
data_fifa_top_1000_wage = data_fifa_top_1000_wage.groupby("Nationality").filter(lambda x: len(x)>=11).head(1000)
data_fifa_top_1000_wage.shape
```

```python
data_fifa_top_1000_wage_avg = data_fifa_top_1000_wage.groupby("Nationality").agg(avg_wage=('wage_clean', 'mean')).sort_values("avg_wage", ascending=False)
data_fifa_top_1000_wage_avg.head()
```

```python
bar1 = px.bar(data_fifa_top_1000_wage_avg.reset_index(), x='Nationality', y='avg_wage',
             title='Средняя зарплата игроков')

bar1.update_yaxes(title_text='Средняя зараплата')

bar1.show()
```

```python
data_fifa_top_1000_wage_count_players = data_fifa_top_1000_wage.groupby("Nationality").agg(players_count=('Nationality', 'count'),avg_wage=('wage_clean', 'mean')).sort_values("players_count", ascending=False)
data_fifa_top_1000_wage_count_players.head()
```

```python
bar1 = px.bar(data_fifa_top_1000_wage_count_players.reset_index(), x='Nationality', y='avg_wage',
              hover_data={'players_count': ':,.2f'},
             title='Количество игроков и средняя зарплата')

bar1.update_yaxes(title_text='Зарплата')

bar1.show()
```
